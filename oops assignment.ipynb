```python
#10 demostrat operator overloading by creating a class vector and overriding the add method to add two vectors
class vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return vector(self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"{self.x}, {self.y}"
    
```
